```python
# Step 2 â€” Import necessary libraries

from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display
import os

print("âœ… Libraries imported successfully!")
```

```python
# Step 3 â€” Take input text and translate it to a target language

# Create translator object
translator = Translator()

# Get text and target language from user
text = input("Enter the text you want to translate: ")
target_lang = input("Enter target language code (e.g., 'fr' for French, 'es' for Spanish, 'hi' for Hindi): ")

# Perform translation
translated = translator.translate(text, dest=target_lang)

# Display the translated text
print("\nğŸŒ Translated Text:")
print(f"({target_lang}) -> {translated.text}")
```

```python
# Improved Step 3 â€” Translate to any language by entering the name (not just code)

from googletrans import LANGUAGES

# Create translator object
translator = Translator()

# Show how many languages are supported
print(f"ğŸŒ Supported languages: {len(LANGUAGES)}")
print("Example inputs: french, spanish, hindi, german, japanese, arabic, etc.\n")

# Take input sentence
text = input("Enter the text you want to translate: ")

# Take input language name (user-friendly)
target_language_name = input("Enter the language you want to translate to: ").lower()

# Find corresponding language code
lang_code = None
for code, name in LANGUAGES.items():
    if name.lower() == target_language_name:
        lang_code = code
        break

if lang_code:
    # Perform translation
    translated = translator.translate(text, dest=lang_code)
    print(f"\nâœ… Translation successful!")
    print(f"Original Text: {text}")
    print(f"Translated ({target_language_name.title()}): {translated.text}")
else:
    print("\nâŒ Sorry, that language is not supported or was misspelled. Please try again.")
```

```python
# Step 4 â€” Convert translated text into speech

from gtts import gTTS
from IPython.display import Audio, display

# Convert the translated text (from Step 3) into audio
if lang_code:
    tts = gTTS(translated.text, lang=lang_code)
    audio_file = "translated_audio.mp3"
    tts.save(audio_file)

    print("\nğŸ”Š Playing translated audio...")
    display(Audio(audio_file, autoplay=True))
else:
    print("âŒ Cannot generate speech. Language code not found.")
```

```python
# Step 5 â€” Combined function for Translation + Text-to-Speech

from googletrans import Translator, LANGUAGES
from gtts import gTTS
from IPython.display import Audio, display

def translate_and_speak():
    translator = Translator()

    print(f"ğŸŒ Supported languages: {len(LANGUAGES)} total")
    print("Example: french, spanish, hindi, german, italian, japanese, etc.\n")

    # Get input text
    text = input("Enter the text you want to translate: ")

    # Get target language name
    target_language_name = input("Enter the language you want to translate to: ").lower()

    # Match the language name to its code
    lang_code = None
    for code, name in LANGUAGES.items():
        if name.lower() == target_language_name:
            lang_code = code
            break

    if not lang_code:
        print("âŒ Language not recognized. Please check spelling.")
        return

    # Perform translation
    translated = translator.translate(text, dest=lang_code)

    print("\nâœ… Translation Successful!")
    print(f"Original Text: {text}")
    print(f"Translated ({target_language_name.title()}): {translated.text}")

    # Convert to speech
    try:
        tts = gTTS(translated.text, lang=lang_code)
        tts.save("translated_output.mp3")
        print("\nğŸ”Š Playing Translated Speech...")
        display(Audio("translated_output.mp3", autoplay=True))
    except Exception as e:
        print("âš ï¸ Could not generate audio for this language. Some languages may not have voice support.")
        print("Error:", e)
```

```python
# Step 6 â€” Add Interactive UI for Cloud-Based Translation + Text-to-Speech

import ipywidgets as widgets
from IPython.display import display, clear_output, Audio
from googletrans import Translator, LANGUAGES
from gtts import gTTS

# Dropdown with language options (sorted alphabetically)
lang_dropdown = widgets.Dropdown(
    options=sorted(LANGUAGES.values()),
    description='ğŸŒ Language:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# Text box for user input
text_input = widgets.Textarea(
    placeholder='Enter text to translate...',
    description='ğŸ“ Input:',
    layout=widgets.Layout(width='600px', height='80px'),
    style={'description_width': 'initial'}
)

# Button to trigger translation
translate_button = widgets.Button(
    description='ğŸ” Translate & Speak',
    button_style='success',
    layout=widgets.Layout(width='250px')
)

# Output area for translated text and audio
output = widgets.Output()

# Define what happens on button click
def on_translate_click(b):
    with output:
        clear_output()
        translator = Translator()
        target_lang_name = lang_dropdown.value
        text = text_input.value

        # Find corresponding language code
        lang_code = None
        for code, name in LANGUAGES.items():
            if name.lower() == target_lang_name.lower():
                lang_code = code
                break

        if not text.strip():
            print("âš ï¸ Please enter some text.")
            return
        if not lang_code:
            print("âŒ Language not recognized.")
            return

        translated = translator.translate(text, dest=lang_code)
        print(f"âœ… Translated ({target_lang_name.title()}):\n{translated.text}")

        # Convert to speech
        try:
            tts = gTTS(translated.text, lang=lang_code)
            tts.save("output.mp3")
            display(Audio("output.mp3", autoplay=True))
        except:
            print("âš ï¸ Audio not available for this language.")

translate_button.on_click(on_translate_click)

# Display UI
display(text_input, lang_dropdown, translate_button, output)
```
